```python
# Dependencies
#from bs4 import BeautifulSoup

import pandas as pd
```

```python
#  website that has all Dow Jones company list
url = "https://www.investopedia.com/terms/d/dow-30.asp"
```

```python
#Scraping the table using Pandas
tables = pd.read_html(url)
```

```python
#Assinging the first second table as DataFrame
df = tables[0]
df.head()
```

```python
#Renaming the columns

df.columns = ['Name', 'Stock Ticker']
df.head()
```

```python
#Removing extra column
df = df.iloc[1:]
df.head()
```

```python
#manually adding facebook, amazon, and Google
df1 = pd.DataFrame([['Amazon', 'AMZN'], ['Google', 'GOOGL'], ['Facebook', 'FB']], columns=list('AB'))
df1.columns = ['Name', 'Stock Ticker']
df1
```

```python
df = df.append(df1)
df
```

```python
df["Headquarters"] = df["Name"] + " Headquarter"
df.head()
```

```python
company_names = []
stock_ticker_names = []
company_headquarters = []

for index, row in df.iterrows():
    company_names.append(row["Name"])
    stock_ticker_names.append(row["Stock Ticker"])
    company_headquarters.append(row["Headquarters"])
```

```python
print(company_names)
print(stock_ticker_names)
print(company_headquarters)
```

```python
#import dependencies

from config import google_api_key
import requests
import gmaps as gmaps
gmaps.configure(api_key=google_api_key)
```

```python
#Making an API call to Google Maps

response = []

for company_headquarter in company_headquarters:

    target_location = company_headquarter 
    
    # Building URL
    target_url = ('https://maps.googleapis.com/maps/api/geocode/json?'
    'address={0}&key={1}').format(target_location, google_api_key)
    
    # Run a request to endpoint and convert result to json
    geo_data = requests.get(target_url).json()
   
    # Add response to our list
    response.append(geo_data)  
```

```python
import pprint as pprint
```

```python
pprint.pprint(response)
```

```python
# Collect and store latitude and longitude for each company headquarter

lat = []
lng = []
i=0


for each in response:
    try:
        latitude = response[i]["results"][0]["geometry"]["location"]["lat"]
        longitude = response[i]["results"][0]["geometry"]["location"]["lng"]

        lat.append(latitude)
        lng.append(longitude)
        
        i = i+1
    except:
        lat.append("lat error")
        lng.append("lng error")
        
        i = i+1
```

```python
df["Latitude"] = lat
df["Longitude"] = lng
df
```
